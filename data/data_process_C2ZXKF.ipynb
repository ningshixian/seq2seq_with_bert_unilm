# 数据预处理

```python
import pandas as pd 
import re
import string
from zhon.hanzi import punctuation

data = pd.read_csv("data/坐席对话数据demo.csv")

data=data[~((data.message.str.contains("转人工") )|(data.message.str.contains("人工坐席")) | (data.message.str.contains("expireDateTime")) | 
            (data.message.str.contains("请问有什么可以帮您")) |(data.message.str.contains("服务进行评价")) |(data.message.str.contains("感谢您的咨询")) 
            | (data.message.str.contains("我是您的人工专属客服")) | (data.message.str.contains("欢迎回来"))
           | (data.message.str.contains("继续为您服务")) | (data.message.str.contains("您是否还在线")) | (data.message.str.contains("请问有什么可以帮助您")) )]
data=data.sort_values("createtime").groupby("contextid").apply(lambda x:x)

def process_text(text):
    # 图片替换为[PIC]
    text = re.sub('(<img[^<]+?>)|(http[s]?://(.*)image(.*))', '[PIC]', str(text)).replace('\n', '')
    # 删掉html标签
    text = re.sub('<[^<]+?>', '', str(text)).replace('\n', '')
    # http替换为[HTTP]
    text = re.sub("http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*,]|(?:%[0-9a-fA-F][0-9a-fA-F]))+", '[HTTP]', text).replace('\t','').replace('&nbsp','').strip()
    
    if text=="[PIC]":
        return text
    
    # 删掉首尾标点
    text = re.sub('^[{}]+|[{}]+$'.format(string.punctuation + punctuation + " ",string.punctuation + punctuation + " "), '', str(text))
    return text
    
data['text'] = data['message'].apply(lambda x: process_text(x))
data = data[~((data.text=="") | (data.text.isin(["你好","您好"])))]

data.to_csv("data/seat/C2ZXKF坐席对话数据20211001至20220301-wwj处理后.csv", index=False)
```

# 数据再加工

```python
import pandas as pd 

data = pd.read_csv("data/seat/C2ZXKF坐席对话数据20211001至20220301-wwj处理后.csv")
```

```python
data['text'] = data[['text', 'calltype']].apply(lambda x: "[KNOW]" if len(str(x['text']))>40 and x['calltype']=="out" else x['text'], axis=1)
```

```python
group_data = data.groupby("contextid")
```

```python
group_list = [x for _, x in group_data if len(x) > 2 and 'out' in x.calltype.values.tolist() and 'in' in x.calltype.values.tolist()]
```

# 划分训练集和测试集

```python
from sklearn.model_selection import train_test_split
#data:需要进行分割的数据集
#random_state:设置随机种子，保证每次运行生成相同的随机数
#test_size:将数据分割成训练集的比例
train_data, test_data = train_test_split(group_list, test_size=0.2, random_state=2022)
```

# 生成测试数据

```python
from tqdm import tqdm
import random

ALL_TEST_LABEL = set()
ALL_TEST_DATA = []

def get_hard_neg_from_qa_pair(qa_pair: list):
    res = []
    if len(res) == 1:
        return res
    index = [x for x in range(len(qa_pair))]
    random.shuffle(index)
    for i in range(0, len(qa_pair)):
        if i != index[i]:
            res.append({'q': qa_pair[i]['q'],
                       'a': qa_pair[index[i]]['a'],
                       'label': 0})
    return res
        
def get_qa_pair_from_df(df):
    global ALL_TEST_LABEL, ALL_TEST_DATA
    res = []
    question = ""
    answer = ""
    answer_html = ""
    for index in df.index:
        if df.loc[index].calltype == "in":
            if answer:
                res.append({"q": question, "a": answer, "label": 1})
                ALL_TEST_LABEL.add((answer_html, answer))
                answer = ""
                answer_html = ""
            if not question:
                question = question + str(df.loc[index].text)
            else:
                question = question + '[SEP]' + str(df.loc[index].text)
        elif df.loc[index].calltype == "out":
            if answer or question:
                if answer and df.loc[index].text != "[KNOW]":
                    answer = answer + "。" + str(df.loc[index].text)
                    answer_html = answer_html + " " + str(df.loc[index].message)
                else:
                    answer = answer + str(df.loc[index].text)
                    answer_html = answer_html + str(df.loc[index].message)
                
    if answer:
        res.append({"q": question, "a": answer, "label": 1})     
    
    res = res + get_hard_neg_from_qa_pair(res)
    return res

for df in tqdm(test_data):
    ALL_TEST_DATA = ALL_TEST_DATA + get_qa_pair_from_df(df)
```

```python
# 随机采样负例，使正例:负例=1:1
true_data = [x for x in ALL_TEST_DATA if x['label'] == 1]
false_data = [x for x in ALL_TEST_DATA if x['label'] == 0]
len_true = len(true_data)
len_false = len(false_data)
all_true_a = [x['a'] for x in true_data]
len_index = [x for x in range(len_true)]
random.shuffle(len_index)
sample_index = len_index[0: (len_true-len_false)]
sample_neg = [{'q': true_data[x]['q'], 'a': random.sample(all_true_a, 1)[0], 'label': 0} for x in sample_index]
ALL_TEST_DATA = ALL_TEST_DATA + sample_neg
```

```python
# 验证正负例数量
print(len([x for x in ALL_TEST_DATA if x['label'] == 1]))
print(len([x for x in ALL_TEST_DATA if x['label'] == 0]))
```
